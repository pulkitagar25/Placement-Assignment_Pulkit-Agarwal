## Q-1. A university wants to understand the relationship between the SAT scores of its applicants and their college GPA. The correlation coefficient of 0.7 between SAT scores and college GPA suggests a strong positive relationship between these two variables in the context of the collected data from the 500 students.

A correlation coefficient measures the strength and direction of the linear relationship between two variables. It ranges from -1 to +1, where values closer to +1 indicate a stronger positive correlation. In this case, the correlation coefficient of 0.7 indicates a strong positive correlation.

Interpreting this correlation coefficient, it implies that as SAT scores increase, there is a tendency for college GPA to increase as well. However, it is important to note that correlation does not imply causation. Other factors, such as individual abilities, study habits, and external influences, may also contribute to the observed relationship.

Therefore, based on the correlation coefficient of 0.7, we can infer that there is a strong positive association between SAT scores and college GPA in the given dataset. Higher SAT scores are generally associated with higher college GPAs among the 500 students studied.

## Q-5. A certain city has two taxi companies: Company A has 80% of the taxis and Company B has 20% of the taxis. Company A's taxis have a 95% success rate for picking up passengers on time, while Company B's taxis have a 90% success rate. If a randomly selected taxi is late, what is the probability that it belongs to Company A?

To calculate the probability that a randomly selected late taxi belongs to Company A, we can use Bayes' theorem. Let's denote the events as follows:

A: Taxi belongs to Company A
B: Taxi belongs to Company B
L: Taxi is late

We are given the following probabilities:

P(A) = 0.8 (Company A has 80% of the taxis)
P(B) = 0.2 (Company B has 20% of the taxis)
P(L|A) = 0.05 (Company A's taxis have a 95% success rate)
P(L|B) = 0.10 (Company B's taxis have a 90% success rate)

We want to calculate P(A|L), the probability that a randomly selected late taxi belongs to Company A.

Using Bayes' theorem:

P(A|L) = (P(A) * P(L|A)) / P(L)

To calculate P(L), the probability of a taxi being late, we can use the law of total probability:

P(L) = P(L|A) * P(A) + P(L|B) * P(B)

Now we can substitute the given values into the equations:

P(L) = (0.05 * 0.8) + (0.10 * 0.2) = 0.04 + 0.02 = 0.06

P(A|L) = (0.8 * 0.05) / 0.06 = 0.04 / 0.06 = 2/3

Therefore, the probability that a randomly selected late taxi belongs to Company A is 2/3 or approximately 0.67.

P(X = k) = (n C k) * p^k * (1 - p)^(n - k)

Where:
P(X = k) is the probability of getting exactly k successes (in this case, defective bulbs)

n is the number of trials (total number of bulbs produced)

k is the number of successes (number of defective bulbs)

p is the probability of success in a single trial (probability of a bulb being defective)

(1 - p) is the probability of failure in a single trial

In this case:
n = 500 (total number of bulbs produced)

k = 20 (number of defective bulbs)

p = 0.05 (probability of a bulb being defective)

Plugging in the values, we can calculate the probability:

P(X = 20) = (500 C 20) * (0.05)^20 * (1 - 0.05)^(500 - 20)

Using a calculator or software, the calculation can be done as follows:

P(X = 20) ≈ 0.1145

Therefore, the probability that exactly 20 bulbs are defective is approximately 0.1145 or 11.45%.
