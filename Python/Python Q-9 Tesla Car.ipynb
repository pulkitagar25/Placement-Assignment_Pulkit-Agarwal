```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
data = pd.read_csv(url)
```

```python
data
```

```python
# Get all cars and their types that do not qualify for clean alternative fuel vehicle
non_clean_cars = data[data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] == "Not eligible due to low battery range"]
non_clean_cars_info = non_clean_cars[["Make", "Model"]]
```

```python
print("Cars and their types that do not qualify for clean alternative fuel vehicle:")
print(non_clean_cars_info)
```

```python
tesla_cars_in_bothell = data[(data["Make"] == "TESLA") & (data["City"] == "Bothell")]
tesla_cars_info = tesla_cars_in_bothell[["Model Year", "Model"]]
```

```python
print("TESLA cars with the model year and model type made in Bothell City:")
print(tesla_cars_info)
```

```python
cars_electric_range_above_100 = data[(data["Electric Range"] > 100) & (data["Model Year"] > 2015)]
print("Cars with an electric range of more than 100 and made after 2015:")
print(cars_electric_range_above_100)
```

```python
# Draw plots to show the distribution between city and electric vehicle type
data["City"] = data["City"].astype(str)
data.plot.scatter(x="City", y="Electric Vehicle Type")
plt.title("Distribution of Electric Vehicle Type by City")
plt.show()
```
